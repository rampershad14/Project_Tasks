```python
import torch
from transformers import BartTokenizer, BartForConditionalGeneration
```

```python
tokenizer = BartTokenizer.from_pretrained("facebook/bart-large-cnn")
model = BartForConditionalGeneration.from_pretrained("facebook/bart-large-cnn")
```

```python
def summarize_text(text):
    # Tokenize the input text
    inputs = tokenizer.encode(text, return_tensors="pt", max_length=1024, truncation=True)

    # Generate summary using the BART model
    summary_ids = model.generate(
        inputs,
        max_length=100,  # Specify max length of the summary
        min_length=30,   # Specify min length of the summary
        length_penalty=2.0,  # Encourage shorter summaries
        num_beams=4,     # Number of beams for beam search
        early_stopping=True
    )

    # Decode the generated summary back into text
    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)

    return summary
```

```python
text = ("Artificial intelligence (AI) is transforming the world we live in. From healthcare to finance, "
        "AI is being applied in various fields to improve efficiency, reduce costs, and provide innovative "
        "solutions. AI-powered systems are helping doctors diagnose diseases more accurately, automating "
        "financial transactions, and even assisting in legal research. However, the widespread use of AI also "
        "raises concerns about data privacy, job displacement, and ethical decision-making. As AI technology "
        "continues to evolve, it will be important to address these challenges to ensure that the benefits are shared by all.")
```

```python
summary = summarize_text(text)
```

```python
print("Original Text:\n", text)
print("\nSummary:\n", summary)
```
