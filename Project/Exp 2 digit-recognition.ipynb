```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist
```

```python
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# 28 px by 28 px images
print("x_train shape is: ", x_train.shape)
print("x_test shape is: ", x_test.shape)
```

```python
# y shows the output of the input, from 0 to 9
print("Shape of y_train:", y_train.shape)
print("Shape of y_test:", y_test.shape)
```

```python
# normalization of the color values (between 0 and 1)
x_train = x_train / 255.0
x_test = x_test / 255.0
```

```python
x_train[57]
```

```python
x_train = x_train.reshape(60000, 784)
x_test = x_test.reshape(10000, 784)
```

```python
# one-hot encoding, useful for softmax activation function
y_train = tf.keras.utils.to_categorical(y_train, 10)
y_test = tf.keras.utils.to_categorical(y_test, 10)
```

```python
y_train[57]
```

```python
print(y_train.shape)
```

```python
# fully connected neural network
model = tf.keras.models.Sequential() # each layer after another
model.add(tf.keras.layers.InputLayer(shape=(784,))) # a single vector of size 28*28 as input
model.add(tf.keras.layers.Dense(128, activation='relu')) #relu brings non-linearlity
model.add(tf.keras.layers.Dense(64, activation='relu'))
model.add(tf.keras.layers.Dense(32, activation='relu'))
model.add(tf.keras.layers.Dense(16, activation='relu'))
model.add(tf.keras.layers.Dense(10, activation='softmax')) #this layer outputs 10 classes, matching y_train
model.compile(optimizer='sgd', loss=tf.keras.losses.categorical_crossentropy, metrics=['accuracy']) # sgd is stochastic gradient descent
model.summary()
model.fit(x_train, y_train, epochs=10, batch_size=10)
```
